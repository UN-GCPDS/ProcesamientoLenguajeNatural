# Introducción a las Expresiones Regulares en Python con la Librería `re`

En este cuaderno, exploraremos el potente mundo de las **Expresiones Regulares (Regex)** en Python, utilizando la librería estándar `re`. Las expresiones regulares son herramientas invaluables para el procesamiento del lenguaje natural (PLN), permitiéndonos buscar, extraer, manipular y validar texto de manera eficiente y flexible.

Comenzaremos con una visión general de las funciones clave que ofrece `re` (como `search`, `match`, `findall`, `sub`, `split` y `compile`), para luego adentrarnos en la creación de **patrones robustos**.

Los patrones robustos son aquellos que no solo buscan coincidencias básicas, sino que también manejan variaciones, extraen información específica y estructurada del texto. Aprenderemos a usar metacaracteres, cuantificadores, clases de caracteres y, muy importante, los **grupos de captura nombrados** (`(?P<nombre>...)`), que nos permitirán acceder a las partes extraídas del texto de una forma mucho más legible y organizada. Estos grupos de captura son esenciales para construir soluciones de PLN que necesitan extraer entidades, fechas, tiempos o cualquier otra pieza de información relevante de grandes volúmenes de texto.

Prepárate para dominar el arte de las expresiones regulares y aplicarlas en tus proyectos de procesamiento de texto.

```python
import re # Librería para trabajar con expresiones regulares
```

```python
texto = """(Jefe de los Minisupers) 14'49'' Podéis hacerme tres preguntas.
(Apu) 14'58'' Qué bueno, porque sólo necesito una.
(Homer) 15'05'' ¿Usted es el Jefe de los Minisupers?
(Jefe de los Minisupers) 15'13'' Así es.
(Homer) 15'22'' ¿Usted?
(Jefe de los Minisupers) 15'38'' Así es.
(Homer) 15'41'' ¿Está seguro?
(Jefe de los Minisupers) 16' Sí. Espero que los haya iluminado.
(Apu) 16'11'' Pero tengo que...
(Jefe de los Minisupers) 16'2'' Gracias. Vuelva pronto.
(Apu) 16'29'' Pero.
(Jefe de los Minisupers) 16'33'' Graciaaaaas. Vuelva pronto."""
```

# Funciones Clave y Uso

*   `re.search(patron, cadena)`: Busca la primera ocurrencia del patrón en toda la cadena.


```python
# Ejemplo de re.search(): Busca la primera aparición del patrón en todo el string.
# Buscamos 'Homer' en el texto.
match_search = re.search(r'Homer', texto)
if match_search:
    print(f"re.search(): 'Homer' encontrado en la posición: {match_search.start()}")
else:
    print("re.search(): 'Homer' no encontrado.")
```

*   `re.match(patron, cadena)`: Intenta hacer coincidir el patrón solo al principio de la cadena.


```python
# Ejemplo de re.match(): Intenta hacer coincidir el patrón solo al principio del string.
# Buscamos '(Jefe de los Minisupers)' al inicio del texto.
match_match = re.match(r'\(Jefe de los Minisupers\)', texto)
if match_match:
    print(f"re.match(): '(Jefe de los Minisupers)' encontrado al inicio: {match_match.group()}")
else:
    print("re.match(): '(Jefe de los Minisupers)' no encontrado al inicio.")

# Un ejemplo que no coincidiría al inicio
match_match_fail = re.match(r'Apu', texto)
if match_match_fail:
    print(f"re.match(): 'Apu' encontrado al inicio: {match_match_fail.group()}")
else:
    print("re.match(): 'Apu' no encontrado al inicio (como se esperaba).")
```

*   `re.findall(patron, cadena)`: Devuelve una lista de todas las coincidencias no superpuestas.


```python
# Ejemplo de re.findall(): Devuelve una lista de todas las coincidencias no superpuestas.
# Buscamos todos los nombres de personajes entre paréntesis.
characters = re.findall(r'\((.*?)\)', texto)
print(f"re.findall(): Nombres de personajes encontrados: {characters}")
```

*   `re.sub(patron, reemplazo, cadena)`: Reemplaza las ocurrencias del patrón.


```python
# Ejemplo de re.sub(): Reemplaza las ocurrencias de un patrón.
# Reemplazamos 'Jefe de los Minisupers' por 'Jefe'.
texto_modificado = re.sub(r'Jefe de los Minisupers', 'Jefe', texto)
print("re.sub(): Texto con 'Jefe de los Minisupers' reemplazado por 'Jefe':")
print(texto_modificado)
```

*   `re.split(patron, cadena)`: Divide la cadena basándose en el patrón.


```python
# Ejemplo de re.split(): Divide el string basándose en el patrón.
# Dividimos el texto por cada nueva línea para obtener las frases individuales.
lines = re.split(r'\n', texto)
print("re.split(): Líneas del diálogo:")
for line in lines:
    print(f"- {line}")
```

*   `re.compile(patron)`: Compila un patrón para su reutilización y mejora del rendimiento.

```python
# Ejemplo de re.compile(): Compila un patrón para su reutilización y rendimiento.
# Compilamos un patrón para buscar tiempos (ej. 14'49'').
pattern_time = re.compile(r'\d{2}\'\d{2}\'\'')

# Usamos el patrón compilado para encontrar todas las ocurrencias.
tiempos = pattern_time.findall(texto)
print(f"re.compile(): Tiempos encontrados usando el patrón compilado: {tiempos}")
```

# Patrones robustos

| Carácter(es) | Qué hace |
|:--------------|:---------|
| `.` | Un punto. Coincide con cualquier carácter excepto el salto de línea. |
| `^` | Un acento circunflejo. Coincide con un patrón al inicio de la cadena. |
| `\A` | A mayúscula. Coincide solo al inicio de la cadena. |
| `$` | Signo de dólar. Coincide con el final de la cadena. |
| `\Z` | Z mayúscula. Coincide solo al final de la cadena. |
| `[ ]` | Coincide con el conjunto de caracteres que se especifican dentro de él. |
| `\` | <ul><li>Si el carácter que sigue a la barra invertida es un carácter de escape reconocido, se toma el significado especial del término.</li><li>De lo contrario, la barra invertida (\\) se trata como cualquier otro carácter y se ignora.</li><li>Se puede usar delante de todos los metacaracteres para eliminar su significado especial.</li></ul> |
| `\w` | w minúscula. Coincide con cualquier letra, dígito o guion bajo. |
| `\W` | W mayúscula. Coincide con cualquier carácter que no sea parte de `\w` (w minúscula). |
| `\s` | s minúscula. Coincide con un solo carácter de espacio en blanco como: espacio, nueva línea, tabulador, retorno. |
| `\S` | S mayúscula. Coincide con cualquier carácter que no sea parte de `\s` (s minúscula). |
| `\d` | d minúscula. Coincide con un dígito decimal 0-9. |
| `\D` | D mayúscula. Coincide con cualquier carácter que no sea un dígito decimal. |
| `\t` | t minúscula. Coincide con un tabulador. |
| `\n` | n minúscula. Coincide con una nueva línea. |
| `\r` | r minúscula. Coincide con un retorno. |
| `\b` | b minúscula. Coincide solo con el principio o el final de la palabra. |
| `+` | Comprueba si el carácter precedente aparece una o más veces. |
| `*` | Comprueba si el carácter precedente aparece cero o más veces. |
| `?` | <ul><li>Comprueba si el carácter precedente aparece exactamente cero o una vez.</li><li>Especifica una versión no codiciosa (menor cantidad de caracteres posible) de `+`, `*`.</li></ul> |
| `{ }` | Comprueba un número explícito de veces. |
| `( )` | Crea un grupo al realizar coincidencias. |
| `< >` | Crea un grupo nombrado al realizar coincidencias. |


## Ejemplo 1

```python
# Patrón robusto con nombres: (?P<nombre_grupo>patron)
patron_avanzado = r"\((?P<personaje>.+?)\)\s+(?P<tiempo>\d{1,2}'(?:\d{0,2}''?)?)\s+(?P<dialogo>.*)"
```

```python
for match in re.finditer(patron_avanzado, texto):
    # Accedemos por nombre, mucho más seguro que usar índices
    print(f"Personaje: {match.group('personaje')} | Dice: {match.group('dialogo')[:20]}...")
```

Aquí la explicación detallada de cada componente del patrón `r"\((?P<personaje>.+?)\)\s+(?P<tiempo>\d{1,2}'(?:\d{0,2}''?)?)\s+(?P<dialogo>.*)"`:

1.  **`\(` y `\)`**: Estos son caracteres especiales en expresiones regulares que denotan grupos de captura. Para que coincidan con los paréntesis literales en el texto `(Jefe de los Minisupers)`, necesitamos 'escaparlos' con una barra invertida (`\`).

2.  **`(?P<personaje>.+?)`**: Esta es la primera parte del patrón que busca el nombre del personaje.
    *   **`(?P<personaje>...)`**: Esto define un **grupo de captura nombrado**. Lo que coincida dentro de los paréntesis se capturará y se le asignará el nombre `personaje`. Esto permite acceder al valor capturado por su nombre en lugar de por un índice numérico, haciendo el código más legible.
    *   **`.`**: Coincide con cualquier carácter (excepto un salto de línea).
    *   **`+`**: Coincide con una o más ocurrencias del carácter anterior (en este caso, cualquier carácter).
    *   **`?` (cuantificador no codicioso o 'non-greedy')**: Cuando se usa después de un cuantificador (`+` o `*`), lo hace *no codicioso*. Esto significa que coincidirá con la menor cantidad de caracteres posible hasta encontrar el siguiente patrón. En este contexto, asegura que `. +?` coincida solo hasta el siguiente `)` literal, evitando que capture más allá de un solo nombre de personaje si hay otros paréntesis más adelante en la línea.
    *   **En resumen, `(?P<personaje>.+?)` captura el nombre del personaje que está entre paréntesis, por ejemplo, "Jefe de los Minisupers" o "Apu".**

3.  **`\s+`**: Coincide con uno o más caracteres de espacio en blanco. Esto se usa para separar el nombre del personaje del siguiente elemento (el tiempo).

4.  **`(?P<tiempo>\d{1,2}'(?:\d{0,2}''?)?)`**: Esta es la segunda parte compleja que captura la marca de tiempo.
    *   **`(?P<tiempo>...)`**: Otro **grupo de captura nombrado**, esta vez llamado `tiempo`.
    *   **`\d{1,2}`**: Coincide con uno o dos dígitos (del 0 al 9). Esto captura la parte de los minutos (ej. `14`, `15`, `16`).
    *   **`'`**: Coincide con el carácter de apóstrofo literal que sigue a los minutos (ej. `14'`)
    *   **`(?:\d{0,2}''?)?`**: Esto es un **grupo no capturador** `(?:...)` que además es **opcional** `?` (el último `?` fuera del grupo).
        *   **`\d{0,2}`**: Dentro del grupo no capturador, coincide con cero, uno o dos dígitos. Esto captura la parte de los segundos (ej. `49`, `05`). Es `0,2` porque a veces los segundos pueden ser de un solo dígito o no estar presentes (como en `16'`).
        *   **`''?`**: Coincide con dos comillas simples (`''`), pero la segunda comilla es opcional (`?`). Esto permite coincidir con `''` (ej. `49''`) o solo con un apóstrofo si no hay segundos (aunque el patrón `\d{1,2}'` ya maneja la parte de minutos y un apóstrofo, esta parte específica se refiere a los segundos y sus comillas). Esto se asegura de que coincida tanto con `14'49''` como con `16'`.
    *   **En resumen, `(?P<tiempo>\d{1,2}'(?:\d{0,2}''?)?)` captura las marcas de tiempo como `14'49''`, `15'05''` o `16'`, y lo nombra 'tiempo'.**

5.  **`\s+`**: Nuevamente, uno o más caracteres de espacio en blanco, separando el tiempo del diálogo.

6.  **`(?P<dialogo>.*)`**: La última parte que captura el diálogo.
    *   **`(?P<dialogo>...)`**: Otro **grupo de captura nombrado**, llamado `dialogo`.
    *   **`.`**: Coincide con cualquier carácter (excepto un salto de línea).
    *   **`*`**: Coincide con cero o más ocurrencias del carácter anterior (en este caso, cualquier carácter).
    *   **En resumen, `(?P<dialogo>.*)` captura todo el texto restante en la línea después del tiempo, por ejemplo, "Podéis hacerme tres preguntas."**

**Propósito general del patrón:**

Este `patron_avanzado` está diseñado para analizar líneas de texto que siguen un formato específico, como las de un diálogo o un log, extrayendo de manera estructurada el nombre del `personaje`, su `tiempo` de intervención y el `dialogo` que pronuncia. El uso de grupos de captura nombrados lo hace especialmente potente y fácil de usar en Python con funciones como `re.match` o `re.finditer` para acceder a la información extraída de forma clara.

## Ejemplo 2

```python
# Otro patrón robusto: Captura personaje, minutos, segundos (opcional) y diálogo.
patron_avanzado_2 = r"\((?P<personaje_nombre>[^)]+?)\)\s+(?P<minutos>\d{1,2})'(?:(?P<segundos>\d{2})''?)?\s+(?P<frase>.*)"

print(f"Nuevo patrón avanzado: {patron_avanzado_2}")

for match in re.finditer(patron_avanzado_2, texto):
    personaje = match.group('personaje_nombre')
    minutos = match.group('minutos')
    segundos = match.group('segundos') if match.group('segundos') else '00'
    frase = match.group('frase')
    print(f"Personaje: {personaje} | Tiempo: {minutos}:{segundos} | Frase: {frase[:30]}...")
```

Este patrón, `r"\((?P<personaje_nombre>[^)]+?)\)\s+(?P<minutos>\d{1,2})'(?:(?P<segundos>\d{2})''?)?\s+(?P<frase>.*)"`, está diseñado para extraer el nombre del personaje, dividir la marca de tiempo en minutos y segundos, y capturar el diálogo de cada línea. Vamos a desglosarlo:

1.  **`\(` y `\)`**: Como en el patrón anterior, `\(` y `\)` son escapes para los paréntesis literales que rodean el nombre del personaje. Si no se escapan, se interpretarían como grupos de captura.

2.  **`(?P<personaje_nombre>[^)]+?)`**: Este es el primer grupo de captura nombrado, llamado `personaje_nombre`.
    *   **`[^)]`**: Coincide con cualquier carácter **excepto** un paréntesis de cierre `)`. Esto es útil para asegurarse de que no se capturen paréntesis internos si los hubiera en el nombre, aunque en nuestro texto no es el caso.
    *   **`+`**: Coincide con una o más ocurrencias del carácter anterior (es decir, cualquier carácter que no sea `)`).
    *   **`?` (cuantificador no codicioso)**: Hace que el `+` sea no codicioso, coincidiendo con la menor cantidad de caracteres posible hasta encontrar el siguiente patrón (el `)` escapado).
    *   **En resumen, `(?P<personaje_nombre>[^)]+?)` captura el nombre del personaje que está dentro de los paréntesis, como "Jefe de los Minisupers" o "Apu".**

3.  **`\s+`**: Coincide con uno o más caracteres de espacio en blanco, separando el nombre del personaje de la marca de tiempo.

4.  **`(?P<minutos>\d{1,2})`**: Este es el grupo de captura nombrado `minutos`.
    *   **`\d{1,2}`**: Coincide con uno o dos dígitos. Esto captura la parte de los minutos de la marca de tiempo (ej. `14`, `15`, `16`).

5.  **`'`**: Coincide con el apóstrofo literal que sigue a los minutos.

6.  **`(?:(?P<segundos>\d{2})''?)?`**: Esta es la parte más compleja que maneja los segundos de la marca de tiempo. Es un grupo **no capturador** `(?:...)` que es **opcional** (el `?` al final).
    *   **`(?P<segundos>\d{2})`**: Dentro del grupo no capturador, hay un grupo de captura nombrado `segundos`.
        *   **`\d{2}`**: Coincide con **exactamente dos dígitos**. Esto está diseñado para capturar los segundos (ej. `49`, `05`). Si los segundos son de un solo dígito o no existen (como en `16'`), este subpatrón no coincidirá y, dado que todo el grupo `(?:...)` es opcional, se permitirá la coincidencia.
    *   **`''?`**: Coincide con dos comillas simples (`''`), donde la segunda es opcional (`?`). Esto permite capturar marcas como `49''` o solo `16'`. Si los segundos no están presentes, este subpatrón tampoco coincidirá.
    *   **En resumen, `(?:(?P<segundos>\d{2})''?)?` intenta capturar dos dígitos como `segundos` seguidos opcionalmente por `''`. Todo el bloque es opcional para manejar casos donde solo hay minutos (ej. `16'`).**

7.  **`\s+`**: Coincide nuevamente con uno o más caracteres de espacio en blanco, separando la marca de tiempo del diálogo.

8.  **`(?P<frase>.*)`**: Este es el último grupo de captura nombrado, llamado `frase`.
    *   **`.`**: Coincide con cualquier carácter (excepto un salto de línea).
    *   **`*`**: Coincide con cero o más ocurrencias del carácter anterior.
    *   **En resumen, `(?P<frase>.*)` captura todo el texto restante en la línea después de la marca de tiempo, considerándolo el diálogo completo.**

**Propósito general de este patrón:**

Este `patron_avanzado_2` es ideal cuando necesitas desglosar la marca de tiempo de manera más granular, obteniendo los minutos y segundos por separado, además del nombre del personaje y su diálogo. Es útil para análisis de tiempo más detallados o para reestructurar los datos en un formato específico (por ejemplo, para convertir el tiempo a un formato `MM:SS`).
